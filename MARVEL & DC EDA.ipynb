```python
#ONLY PROTAGINIST
good = marvel[marvel['Alignment']=='good']
good
```

```python
#ONLY ANTAGONISTS
bad = marvel[marvel['Alignment']=='bad']
bad
```

```python
#GOOD CHARACTERS ACCORDING TO THEIR SPEED
good.sort_values(by = 'Speed', ascending = False)
```

```python
#BAD CHARACTERS ACCORDING TO THEIR SPEED
bad.sort_values(by = 'Speed',ascending = False)
```

```python
#GOOD CHARACTERS ACCORDING THEIR POWER
good.sort_values( by = 'Power' , ascending = False)
```

```python
good.sort_values( by = 'Power' , ascending = False).head()
```

```python
#PLOTTING A BAR-PLOT OF TOP 5 MOST POWER CHARACTERS
sea.set(style = 'whitegrid')
sea.barplot (x= 'Name', y = 'Power' , data =good.sort_values( by = 'Power' , ascending = False).head() )
```
